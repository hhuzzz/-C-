```python
# 划分训练集和测试集
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.3, random_state=42
)

# 定义模型
# model_1 = DecisionTreeRegressor(random_state=42)
model_1 = xgb.XGBRegressor(
    n_estimators=100, max_depth=10, learning_rate=0.1, random_state=42
)
model_2 = RandomForestRegressor(random_state=42)
model_3 = GradientBoostingRegressor(max_depth=10, random_state=42)
# model_3 = GradientBoostingRegressor(random_state=42)

# 模型训练
model_1.fit(X_train, Y_train)
model_2.fit(X_train, Y_train)
model_3.fit(X_train, Y_train)

# 模型预测
Y_pred_1 = model_1.predict(X_test)
Y_pred_2 = model_2.predict(X_test)
Y_pred_3 = model_3.predict(X_test)


```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. 预测值 vs 实际值的散点图
def plot_predictions(Y_test, Y_pred_1, Y_pred_2, Y_pred_3):
    plt.figure(figsize=(15, 5))

    # XGBoost 预测结果
    plt.subplot(1, 3, 1)
    plt.scatter(Y_test, Y_pred_1, color='blue', edgecolor='k', alpha=0.6, label='XGBoost')
    plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'k--', lw=2)
    plt.xlabel('实际值')
    plt.ylabel('预测值')
    plt.title('XGBoost: 预测值 vs 实际值')
    plt.legend()

    # 随机森林 预测结果
    plt.subplot(1, 3, 2)
    plt.scatter(Y_test, Y_pred_2, color='green', edgecolor='k', alpha=0.6, label='随机森林')
    plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'k--', lw=2)
    plt.xlabel('实际值')
    plt.ylabel('预测值')
    plt.title('随机森林: 预测值 vs 实际值')
    plt.legend()

    # GBDT 预测结果
    plt.subplot(1, 3, 3)
    plt.scatter(Y_test, Y_pred_3, color='red', edgecolor='k', alpha=0.6, label='GBDT')
    plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'k--', lw=2)
    plt.xlabel('实际值')
    plt.ylabel('预测值')
    plt.title('GBDT: 预测值 vs 实际值')
    plt.legend()

    plt.tight_layout()
    plt.show()

# 2. 残差图
def plot_residuals(Y_test, Y_pred_1, Y_pred_2, Y_pred_3):
    residuals_1 = Y_test - Y_pred_1
    residuals_2 = Y_test - Y_pred_2
    residuals_3 = Y_test - Y_pred_3

    plt.figure(figsize=(15, 5))

    # XGBoost 残差
    plt.subplot(1, 3, 1)
    plt.scatter(Y_test, residuals_1, color='blue', edgecolor='k', alpha=0.6, label='XGBoost 残差')
    plt.axhline(y=0, color='k', linestyle='--', lw=2)
    plt.xlabel('实际值')
    plt.ylabel('残差')
    plt.title('XGBoost 残差图')
    plt.legend()

    # 随机森林 残差
    plt.subplot(1, 3, 2)
    plt.scatter(Y_test, residuals_2, color='green', edgecolor='k', alpha=0.6, label='随机森林 残差')
    plt.axhline(y=0, color='k', linestyle='--', lw=2)
    plt.xlabel('实际值')
    plt.ylabel('残差')
    plt.title('随机森林 残差图')
    plt.legend()

    # GBDT 残差
    plt.subplot(1, 3, 3)
    plt.scatter(Y_test, residuals_3, color='red', edgecolor='k', alpha=0.6, label='GBDT 残差')
    plt.axhline(y=0, color='k', linestyle='--', lw=2)
    plt.xlabel('实际值')
    plt.ylabel('残差')
    plt.title('GBDT 残差图')
    plt.legend()

    plt.tight_layout()
    plt.show()

# 3. 特征重要性图
def plot_feature_importance(model_1, model_2, model_3, feature_names):
    plt.figure(figsize=(15, 5))

    # XGBoost 特征重要性
    plt.subplot(1, 3, 1)
    plt.barh(feature_names, model_1.feature_importances_, color='blue', alpha=0.6)
    plt.xlabel('重要性得分')
    plt.title('XGBoost 特征重要性')

    # 随机森林 特征重要性
    plt.subplot(1, 3, 2)
    plt.barh(feature_names, model_2.feature_importances_, color='green', alpha=0.6)
    plt.xlabel('重要性得分')
    plt.title('随机森林 特征重要性')

    # GBDT 特征重要性
    plt.subplot(1, 3, 3)
    plt.barh(feature_names, model_3.feature_importances_, color='red', alpha=0.6)
    plt.xlabel('重要性得分')
    plt.title('GBDT 特征重要性')

    plt.tight_layout()
    plt.show()

# 调用绘图函数
plot_predictions(Y_test, Y_pred_1, Y_pred_2, Y_pred_3)
plot_residuals(Y_test, Y_pred_1, Y_pred_2, Y_pred_3)

# 假设 X_train 有列名
feature_names = X_train.columns
plot_feature_importance(model_1, model_2, model_3, feature_names)
```

```python
# 网格搜索最佳比例
# 自定义加权平均函数
def weighted_ensemble_predict(weights, Y_pred_1, Y_pred_2, Y_pred_3):
    return weights[0] * Y_pred_1 + weights[1] * Y_pred_2 + weights[2] * Y_pred_3


# 定义损失函数（这里用绝对误差）
def loss_function(weights):
    Y_pred_ensemble = weighted_ensemble_predict(weights, Y_pred_1, Y_pred_2, Y_pred_3)
    return mean_absolute_error(Y_test, Y_pred_ensemble)


# 网格搜索权重
param_grid = {
    "weights": [
        [0.1, 0.8, 0.1],
        [0.2, 0.6, 0.2],
        [0.3, 0.4, 0.3],
        [0.4, 0.2, 0.4],
        [0.5, 0.2, 0.3],
        [0.6, 0.2, 0.2],
        [0.7, 0.15, 0.15],
        [0.8, 0.1, 0.1],
        [0.9, 0.05, 0.05],
        [1.0, 0, 0],
    ]
}

# 使用 GridSearchCV 搜索最佳权重
best_weights = None
best_score = float("inf")

for weights in param_grid["weights"]:
    score = loss_function(weights)
    if score < best_score:
        best_score = score
        best_weights = weights

print("最佳权重组合:", best_weights)
print("对应的最小绝对误差:", best_score)

# 保存每个权重组合对应的均方误差
errors = []

# 计算每个权重组合的均方误差
for weights in param_grid["weights"]:
    score = loss_function(weights)  # 使用之前定义的 loss_function 计算误差
    errors.append(score)

# 可视化权重组合与误差
weights_array = np.array(param_grid["weights"])
errors_array = np.array(errors)

# 散点图展示
plt.figure(figsize=(10, 6))
plt.scatter(np.arange(len(errors)), errors_array, color="b", label="MAE")

# 给每个点标注权重组合
for i, txt in enumerate(param_grid["weights"]):
    plt.annotate(txt, (i, errors_array[i]))

plt.xlabel("权重组合索引")
plt.ylabel("绝对误差 (MAE)")
plt.title("不同权重组合下的绝对误差")
plt.legend()
plt.grid(True)
plt.show()

# 使用最佳权重进行最终预测
Y_pred_ensemble = weighted_ensemble_predict(best_weights, Y_pred_1, Y_pred_2, Y_pred_3)

# 集成：加权平均
# Y_pred_ensemble = Y_pred_1 * 0.3 + Y_pred_2 * 0.4 + Y_pred_3 * 0.3

# 计算单个模型的评估指标
r2_1 = r2_score(Y_test, Y_pred_1)
r2_2 = r2_score(Y_test, Y_pred_2)
r2_3 = r2_score(Y_test, Y_pred_3)
r2_ensemble = r2_score(Y_test, Y_pred_ensemble)

mape_1 = mean_absolute_percentage_error(Y_test, Y_pred_1)
mape_2 = mean_absolute_percentage_error(Y_test, Y_pred_2)
mape_3 = mean_absolute_percentage_error(Y_test, Y_pred_3)
mape_ensemble = mean_absolute_percentage_error(Y_test, Y_pred_ensemble)

mae_1 = mean_absolute_error(Y_test, Y_pred_1)
mae_2 = mean_absolute_error(Y_test, Y_pred_2)
mae_3 = mean_absolute_error(Y_test, Y_pred_3)
mae_ensemble = mean_absolute_error(Y_test, Y_pred_ensemble)

mse_1 = mean_squared_error(Y_test, Y_pred_1)
mse_2 = mean_squared_error(Y_test, Y_pred_2)
mse_3 = mean_squared_error(Y_test, Y_pred_3)
mse_ensemble = mean_squared_error(Y_test, Y_pred_ensemble)

# 输出评估指标
print(f"XGBoost R²: {r2_1:.4f}, MAPE: {mape_1:.4f},MAE: {mae_1:.4f}, MSE: {mse_1:.4f}")
print(
    f"RandomForest R²: {r2_2:.4f}, MAPE: {mape_2:.4f}, MAE: {mae_2:.4f}, MSE: {mse_2:.4f}"
)
print(
    f"GradientBoosting R²: {r2_3:.4f}, MAPE: {mape_3:.4f}, MAE: {mae_3:.4f}, MSE: {mse_3:.4f}"
)
print(
    f"集成模型 R²: {r2_ensemble:.4f}, MAPE: {mape_ensemble:.4f}, MAE: {mae_ensemble:.4f}, MSE: {mse_ensemble:.4f}"
)

print("预测值的平均绝对差异：")
print(f"XGBoost vs Random Forest: {np.mean(np.abs(Y_pred_1 - Y_pred_2)):.4f}")
print(f"XGBoost vs Gradient Boosting: {np.mean(np.abs(Y_pred_1 - Y_pred_3)):.4f}")
print(f"Random Forest vs Gradient Boosting: {np.mean(np.abs(Y_pred_2 - Y_pred_3)):.4f}")
```

```python
# 绘制 R² 和 MAPE 的柱状图
models = ["XGBoost", "RandomForest", "GradientBoosting", "Ensemble"]
r2_scores = [r2_1, r2_2, r2_3, r2_ensemble]
mape_scores = [mape_1, mape_2, mape_3, mape_ensemble]
mae_scores = [mae_1, mae_2, mae_3, mae_ensemble]

# 绘制 R² 的柱状图
plt.figure(figsize=(10, 5))
plt.bar(models, r2_scores, color=["blue", "green", "orange", "red"])
plt.title(r"R$^2$ Scores for Different Models")
plt.xlabel("Models")
plt.ylabel(r"R$^2$ Score")
plt.show()

# 绘制 MAPE 的柱状图
plt.figure(figsize=(10, 5))
plt.bar(models, mape_scores, color=["blue", "green", "orange", "red"])
plt.title("MAPE Scores for Different Models")
plt.xlabel("Models")
plt.ylabel("MAPE")
plt.show()

# 绘制MAE 的柱状图
plt.figure(figsize=(10, 5))
plt.bar(models, mae_scores, color=["blue", "green", "orange", "red"])
plt.title("MAE Scores for Different Models")
plt.xlabel("Models")
plt.ylabel("MAE")
plt.show()
```

```python

def plot_compare(Y_pred):
    # 绘制预测值与真实值对比图
    plt.figure(figsize=(6, 4))
    plt.scatter(Y_test, Y_pred, alpha=0.7, color='b')
    plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], color='r', linestyle='--')  # 理想情况的对角线
    plt.title(f' 预测值 vs 真实值')
    plt.xlabel('真实值')
    plt.ylabel('预测值')
    plt.show()

    # 绘制残差图（真实值 - 预测值）
    plt.figure(figsize=(6, 4))
    sns.histplot(Y_test - Y_pred, kde=True, color='g', bins=30, alpha=0.6)
    plt.title(f'残差分布图')
    plt.xlabel('残差')
    plt.ylabel('频率')
    plt.show()

    # 绘制实际值和预测值的对比折线图
    plt.figure(figsize=(10, 6))
    plt.plot(range(len(Y_test)), Y_test, label='真实值', color='b', marker='o', alpha=0.6)
    plt.plot(range(len(Y_pred)), Y_pred, label='预测值', color='r', linestyle='--', marker='x', alpha=0.6)
    plt.title(f' 真实值与预测值对比')
    plt.xlabel('样本索引')
    plt.ylabel('值')
    plt.legend()
    plt.show()

plot_compare(Y_pred_1)
plot_compare(Y_pred_2)
plot_compare(Y_pred_3)
plot_compare(Y_pred_ensemble)
```

```python
def plot_compare(Y_pred, model_name):
    # 创建一个图像窗口，包含 1 行 3 列的子图
    fig, axs = plt.subplots(1, 3, figsize=(18, 5))

    # 1. 预测值 vs 真实值的散点图
    axs[0].scatter(Y_test, Y_pred, alpha=0.7, color='b')
    axs[0].plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], color='r', linestyle='--')  # 理想情况的对角线
    axs[0].set_title(f'{model_name} 预测值 vs 真实值')
    axs[0].set_xlabel('真实值')
    axs[0].set_ylabel('预测值')

    # 2. 残差分布图
    sns.histplot(Y_test - Y_pred, kde=True, color='g', bins=30, alpha=0.6, ax=axs[1])
    axs[1].set_title(f'{model_name} 残差分布图')
    axs[1].set_xlabel('残差')
    axs[1].set_ylabel('频率')

    # 3. 真实值与预测值对比折线图
    axs[2].plot(range(len(Y_test)), Y_test, label='真实值', color='b', marker='o', alpha=0.6)
    axs[2].plot(range(len(Y_pred)), Y_pred, label='预测值', color='r', linestyle='--', marker='x', alpha=0.6)
    axs[2].set_title(f'{model_name} 真实值与预测值对比')
    axs[2].set_xlabel('样本索引')
    axs[2].set_ylabel('值')
    axs[2].legend()

    # 调整布局
    plt.tight_layout()
    plt.show()

# 对每个模型绘制综合图
plot_compare(Y_pred_1, "XGBoost")
plot_compare(Y_pred_2, "随机森林")
plot_compare(Y_pred_3, "GBDT")
plot_compare(Y_pred_ensemble, "加权集成模型")
```
