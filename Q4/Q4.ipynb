```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
import xgboost as xgb
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.sans-serif']=['SimHei'] # 中文支持
plt.rcParams['axes.unicode_minus']=False

def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100


```

### K-Fold cross validation of XGBoost, Random Forest, and GBDT

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import KFold, cross_val_predict
from sklearn.metrics import r2_score, mean_absolute_percentage_error, mean_squared_error, mean_absolute_error
import numpy as np

# 假设 Y 是损耗列
X = pd.concat(
    [
        pd.get_dummies(data[["磁芯材料", "温度，oC", "励磁波形"]]).reset_index(drop=True),
        data[["频率，Hz", "磁通密度峰值", "峰峰值"]].reset_index(drop=True),
    ],
    axis=1,
) 

# 定义模型
# model_1 = DecisionTreeRegressor(random_state=42)
model_1 = xgb.XGBRegressor(n_estimators=100, max_depth=10, learning_rate=0.1, random_state=42)
model_2 = RandomForestRegressor(random_state=42)
model_3 = GradientBoostingRegressor(
    max_depth=10, random_state=42
)

# 设置K折交叉验证
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# 预测结果使用cross_val_predict生成
Y_pred_1 = cross_val_predict(model_1, X, Y, cv=kf)
Y_pred_2 = cross_val_predict(model_2, X, Y, cv=kf)
Y_pred_3 = cross_val_predict(model_3, X, Y, cv=kf)

# 计算评估指标
r2_xgb = r2_score(Y, Y_pred_1)
mape_xgb = mean_absolute_percentage_error(Y, Y_pred_1)
mse_xgb = mean_squared_error(Y, Y_pred_1)
mae_xgb = mean_absolute_error(Y, Y_pred_1)

r2_rf = r2_score(Y, Y_pred_2)
mape_rf = mean_absolute_percentage_error(Y, Y_pred_2)
mse_rf = mean_squared_error(Y, Y_pred_2)
mae_rf = mean_absolute_error(Y, Y_pred_2)

r2_gbdt = r2_score(Y, Y_pred_3)
mape_gbdt = mean_absolute_percentage_error(Y, Y_pred_3)
mse_gbdt = mean_squared_error(Y, Y_pred_3)
mae_gbdt = mean_absolute_error(Y, Y_pred_3)

# 进行加权平均的集成学习（可以调整权重）
# Y_pred_ensemble = (Y_pred_1 * 0.5 + Y_pred_2 * 0.5 + Y_pred_3 * 0)
Y_pred_ensemble = (Y_pred_1 * 0.8 + Y_pred_2 * 0.1 + Y_pred_3 * 0.1)
Y_r2_ensemble = r2_score(Y, Y_pred_ensemble)
Y_mape_ensemble = mean_absolute_percentage_error(Y, Y_pred_ensemble)
Y_mse_ensemble = mean_squared_error(Y, Y_pred_ensemble)
Y_mae_ensemble = mean_absolute_error(Y, Y_pred_ensemble)

# print(f"K折交叉验证 集成模型 R² Score: {r2:.4f}")
# print(f"K折交叉验证 集成模型 MAPE: {mape:.4f}")
# print(f"K折交叉验证 集成模型 MSE: {mse:.4f}")
# print(f"K折交叉验证 集成模型 MAE: {mae:.4f}")
```

```python
# 输出评估指标
print(f"XGBoost R²: {r2_xgb:.4f}, MAPE: {mape_xgb:.4f},MAE: {mae_xgb:.4f}, MSE: {mse_xgb:.4f}")
print(
    f"RandomForest R²: {r2_rf:.4f}, MAPE: {mape_rf:.4f}, MAE: {mae_rf:.4f}, MSE: {mse_rf:.4f}"
)
print(
    f"GradientBoosting R²: {r2_gbdt:.4f}, MAPE: {mape_gbdt:.4f}, MAE: {mae_gbdt:.4f}, MSE: {mse_gbdt:.4f}"
)
print(
    f"集成模型 R²: {Y_r2_ensemble:.4f}, MAPE: {Y_mape_ensemble:.4f}, MAE: {Y_mae_ensemble:.4f}, MSE: {Y_mse_ensemble:.4f}"
)
```

```python

# 绘制 R² 和 MAPE 的柱状图
models = ["XGBoost", "RandomForest", "GBDT", "Ensemble"]
r2_scores = [r2_xgb, r2_rf, r2_gbdt, Y_r2_ensemble]
mape_scores = [mape_xgb, mape_rf, mape_gbdt, Y_mape_ensemble]
mae_scores = [mae_xgb, mae_rf, mae_gbdt, Y_mae_ensemble]
mse_scores = [mse_xgb, mse_rf, mse_gbdt, Y_mse_ensemble]

# 绘制 R² 的柱状图
plt.figure(figsize=(10, 5))
plt.bar(models, r2_scores, color=["blue", "green", "orange", "red"])
plt.title(r"R$^2$ Scores for Different Models")
plt.xlabel("Models")
plt.ylabel(r"R$^2$ Score")
plt.show()

# 绘制 MAPE 的柱状图
plt.figure(figsize=(10, 5))
plt.bar(models, mape_scores, color=["blue", "green", "orange", "red"])
plt.title("MAPE Scores for Different Models")
plt.xlabel("Models")
plt.ylabel("MAPE")
plt.show()

# 绘制MAE 的柱状图
plt.figure(figsize=(10, 5))
plt.bar(models, mae_scores, color=["blue", "green", "orange", "red"])
plt.title("MAE Scores for Different Models")
plt.xlabel("Models")
plt.ylabel("MAE")
plt.show()

# 绘制MSE 的柱状图
plt.figure(figsize=(10, 5))
plt.bar(models, mse_scores, color=["blue", "green", "orange", "red"])
plt.title("MSE Scores for Different Models")
plt.xlabel("Models")
plt.ylabel("MSE")
plt.show()
```

```python
# 划分训练集和测试集
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.3, random_state=42
)

# 定义模型
# model_1 = DecisionTreeRegressor(random_state=42)
model_1 = xgb.XGBRegressor(
    n_estimators=100, max_depth=10, learning_rate=0.1, random_state=42
)
model_2 = RandomForestRegressor(random_state=42)
model_3 = GradientBoostingRegressor(max_depth=10, random_state=42)
# model_3 = GradientBoostingRegressor(random_state=42)

# 模型训练
model_1.fit(X_train, Y_train)
model_2.fit(X_train, Y_train)
model_3.fit(X_train, Y_train)

# 模型预测
Y_pred_1 = model_1.predict(X_test)
Y_pred_2 = model_2.predict(X_test)
Y_pred_3 = model_3.predict(X_test)
```

```python
# 网格搜索最佳比例
# 自定义加权平均函数
def weighted_ensemble_predict(weights, Y_pred_1, Y_pred_2, Y_pred_3):
    return weights[0] * Y_pred_1 + weights[1] * Y_pred_2 + weights[2] * Y_pred_3


# 定义损失函数（这里用绝对误差）
def loss_function(weights):
    Y_pred_ensemble = weighted_ensemble_predict(weights, Y_pred_1, Y_pred_2, Y_pred_3)
    return mean_absolute_error(Y_test, Y_pred_ensemble)


# 网格搜索权重
param_grid = {
    "weights": [
        [0.1, 0.8, 0.1],
        [0.2, 0.6, 0.2],
        [0.3, 0.4, 0.3],
        [0.4, 0.2, 0.4],
        [0.5, 0.2, 0.3],
        [0.6, 0.2, 0.2],
        [0.7, 0.15, 0.15],
        [0.8, 0.1, 0.1],
        [0.9, 0.05, 0.05],
        [1.0, 0, 0]
        # 可以继续添加更多组合
    ]
}

# 使用 GridSearchCV 搜索最佳权重
best_weights = None
best_score = float("inf")

for weights in param_grid["weights"]:
    score = loss_function(weights)
    if score < best_score:
        best_score = score
        best_weights = weights

print("最佳权重组合:", best_weights)
print("对应的最小绝对误差:", best_score)

# 保存每个权重组合对应的均方误差
errors = []

# 计算每个权重组合的均方误差
for weights in param_grid["weights"]:
    score = loss_function(weights)  # 使用之前定义的 loss_function 计算误差
    errors.append(score)

# 可视化权重组合与误差
weights_array = np.array(param_grid["weights"])
errors_array = np.array(errors)

# 散点图展示
plt.figure(figsize=(10, 6))
plt.scatter(np.arange(len(errors)), errors_array, color="b", label="MAE")

# 给每个点标注权重组合
for i, txt in enumerate(param_grid["weights"]):
    plt.annotate(txt, (i, errors_array[i]))

plt.xlabel("权重组合索引")
plt.ylabel("均方误差 (MAE)")
plt.title("不同权重组合下的均方误差")
plt.legend()
plt.grid(True)
plt.show()

# 使用最佳权重进行最终预测
Y_pred_ensemble = weighted_ensemble_predict(best_weights, Y_pred_1, Y_pred_2, Y_pred_3)

# 集成：加权平均
# Y_pred_ensemble = Y_pred_1 * 0.3 + Y_pred_2 * 0.4 + Y_pred_3 * 0.3

# 计算单个模型的评估指标
r2_1 = r2_score(Y_test, Y_pred_1)
r2_2 = r2_score(Y_test, Y_pred_2)
r2_3 = r2_score(Y_test, Y_pred_3)
r2_ensemble = r2_score(Y_test, Y_pred_ensemble)

mape_1 = mean_absolute_percentage_error(Y_test, Y_pred_1)
mape_2 = mean_absolute_percentage_error(Y_test, Y_pred_2)
mape_3 = mean_absolute_percentage_error(Y_test, Y_pred_3)
mape_ensemble = mean_absolute_percentage_error(Y_test, Y_pred_ensemble)

mae_1 = mean_absolute_error(Y_test, Y_pred_1)
mae_2 = mean_absolute_error(Y_test, Y_pred_2)
mae_3 = mean_absolute_error(Y_test, Y_pred_3)
mae_ensemble = mean_absolute_error(Y_test, Y_pred_ensemble)


# 输出评估指标
print(f"XGBoost R²: {r2_1:.4f}, MAPE: {mape_1:.4f},MAE: {mae_1:.4f}")
print(f"RandomForest R²: {r2_2:.4f}, MAPE: {mape_2:.4f}, MAE: {mae_2:.4f}")
print(f"GradientBoosting R²: {r2_3:.4f}, MAPE: {mape_3:.4f}, MAE: {mae_3:.4f}")
print(f"集成模型 R²: {r2_ensemble:.4f}, MAPE: {mape_ensemble:.4f}, MAE: {mae_ensemble:.4f}")

print("预测值的平均绝对差异：")
print(f"XGBoost vs Random Forest: {np.mean(np.abs(Y_pred_1 - Y_pred_2)):.4f}")
print(f"XGBoost vs Gradient Boosting: {np.mean(np.abs(Y_pred_1 - Y_pred_3)):.4f}")
print(f"Random Forest vs Gradient Boosting: {np.mean(np.abs(Y_pred_2 - Y_pred_3)):.4f}")
```

```python
# 绘制 R² 和 MAPE 的柱状图
models = ["XGBoost", "RandomForest", "GradientBoosting", "Ensemble"]
r2_scores = [r2_1, r2_2, r2_3, r2_ensemble]
mape_scores = [mape_1, mape_2, mape_3, mape_ensemble]
mae_scores = [mae_1, mae_2, mae_3, mae_ensemble]

# 绘制 R² 的柱状图
plt.figure(figsize=(10, 5))
plt.bar(models, r2_scores, color=["blue", "green", "orange", "red"])
plt.title(r"R$^2$ Scores for Different Models")
plt.xlabel("Models")
plt.ylabel(r"R$^2$ Score")
plt.show()

# 绘制 MAPE 的柱状图
plt.figure(figsize=(10, 5))
plt.bar(models, mape_scores, color=["blue", "green", "orange", "red"])
plt.title("MAPE Scores for Different Models")
plt.xlabel("Models")
plt.ylabel("MAPE")
plt.show()

# 绘制MAE 的柱状图
plt.figure(figsize=(10, 5))
plt.bar(models, mae_scores, color=["blue", "green", "orange", "red"])
plt.title("MAE Scores for Different Models")
plt.xlabel("Models")
plt.ylabel("MAE")
plt.show()
```

```python

def plot_compare(Y_pred):
    # 绘制预测值与真实值对比图
    plt.figure(figsize=(6, 4))
    plt.scatter(Y_test, Y_pred, alpha=0.7, color='b')
    plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], color='r', linestyle='--')  # 理想情况的对角线
    plt.title(f' 预测值 vs 真实值')
    plt.xlabel('真实值')
    plt.ylabel('预测值')
    plt.show()

    # 绘制残差图（真实值 - 预测值）
    plt.figure(figsize=(6, 4))
    sns.histplot(Y_test - Y_pred, kde=True, color='g', bins=30, alpha=0.6)
    plt.title(f'残差分布图')
    plt.xlabel('残差')
    plt.ylabel('频率')
    plt.show()

    # 绘制实际值和预测值的对比折线图
    plt.figure(figsize=(10, 6))
    plt.plot(range(len(Y_test)), Y_test, label='真实值', color='b', marker='o', alpha=0.6)
    plt.plot(range(len(Y_pred)), Y_pred, label='预测值', color='r', linestyle='--', marker='x', alpha=0.6)
    plt.title(f' 真实值与预测值对比')
    plt.xlabel('样本索引')
    plt.ylabel('值')
    plt.legend()
    plt.show()

plot_compare(Y_pred_1)
plot_compare(Y_pred_2)
plot_compare(Y_pred_3)
plot_compare(Y_pred_ensemble)
```
